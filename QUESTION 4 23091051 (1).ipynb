```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(20)


coefficients = np.random.randint(-10, 10, 6)  


def polynomial(x):
    return np.polyval(coefficients, x)


x = np.linspace(-5, 5, 1000)


plt.figure(figsize=(10, 6))
plt.plot(x, polynomial(x), label='Polynomial')
deriv = np.polyder(coefficients)
roots = np.roots(deriv)
extrema = [x for x in roots if x.imag == 0 and -5 <= x.real <= 5]
extrema_values = polynomial(extrema)

for i, (ex, ey) in enumerate(zip(extrema, extrema_values)):
    label = f'Maxima {i+1}' if deriv[5-i] < 0 else f'Minima {i+1}'
    plt.annotate(label, (ex, ey), textcoords="offset points", xytext=(0,10), ha='center')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of a 5th Degree Polynomial with Maxima and Minima')
plt.legend()
plt.grid(True)
plt.show()
```
